```python
import random
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches
```

```python
x = random.random()
y = random.random()

theta = np.linspace( 0 , 2 * np.pi , 150 )

radius = 1/2

a = radius * np.cos( theta )+1/2
b = radius * np.sin( theta )+1/2

figure, axes = plt.subplots( 1 )

#axes.plot( a, b )
axes.set_aspect( 1 )

plt.scatter(x,y)
plt.xlim((0,1))
plt.ylim((0,1))
plt.plot(a,b)
```
